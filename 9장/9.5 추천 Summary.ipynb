<h1>추천 Summary</h1>

<h2>추천 시스템의 중요성</h2>

지금까지 학습한 추천시스템에 대해 간략하게 요약해보자.

무수의 E-commerce 업체들이 이 추천 시스템을 고도화해서 상품의 매출을 어떻게 올릴 것이냐.

특히나 온라인 업체들은 매우 많은 상품들이 있기 때문에 이 상품을 어떻게 최적화를 하여 추천할 것이냐에 따라서

매출에 기여에 많은 영향을 미친다.

특히 엔터테인먼트 업체 같은. 유튜브라던가. 이런 곳은 조금이라도 사용자들을 자신의 사이트에 끌어들이기 위해서 노력한다.

일반적으로 **추천시스템의 유형**은 두가지로 나뉜다.

<h3>1. 콘텐츠 기반 필터링</h3>

상품의 아이템이 가지고 있는 여러가지 요소들 즉 컨텐츠들(장르라던지 description이 됐던지. 가령 영화를 예로 들어보면 감독이나 주연배우나 또는 영화가 가지고 있는 여러가지 요소들의 유사도에 따라서 비슷한 컨텐츠 적 유사도를 가진 영화를 추천하는 그런 방식이다.

<h3>2. 협업 필터링</h3>

아는 사람에게 물어봐라. 그래서 내가 보거나 구매하지 않은 상품이나 아이템 들에 대해서 그것을 이미 구매하거나 본 사람에게 받는 것.

협업 필터링은 다시 두가지로 나뉘는데

**-최근접 이웃기반(Nearest Neighbor)**(친구에게 물어봐라.)<br>
사용자 기반 (User-user CF): 나와 비슷한 취향을 가진 다른 사람을 찾아서 그 사람이 선호하는 아이템을 추천하는 방식<br>
아이템 기반 (Item-item CF): 사용자에게 비슷한 유형의 평점을 받은, 비슷한 유형의 선호가 되었던 그런 상품들에 대해서 비슷한 유형의 평점을 받은 다른 상품을 추천하는 방식. 보통 이게 더 사용자 기반보다 선호 된다. 그 이유로써는 사람이 복잡 다양한 면이 있기 떄문에 상품을 비슷한거를 선택했다고해서 그 사람이 취향이 같냐고 말하기에는 취약점이 있다.


**-잠재 요인 기반(Latent Factor)**<br>
행렬 분해 기반(Matrix Factorization): 숨겨져있던 Latent Factor를 찾아서 재결합을 하여 구매하지 않은 상품, 아직 보지 않은 영화에 대한 평점을 예측하는 것이다.


각각의 방법에 대해서 직접 코드를 작성해봤으나 보통은 패키지가 있다.
